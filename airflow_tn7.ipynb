```python
# reshaping data 
# import necessary libraries
import os
import glob
  
  
# use glob to get all the csv files 
# in the folder
path = r'/content/data_skenario'
csv_files = glob.glob(os.path.join(path, "*.xlsx"))

array_ = []
# loop over the list of csv files
for f in csv_files:
    
    df = pd.read_excel(f)
    df.pop('Temp')
    array_.append(df)
```

```python
if not os.path.exists("images"):
    os.mkdir("images")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plot_array = []
for i in range(len(array_)):
  x = np.array(array_[i]['koor_x'])
  y = np.array(array_[i]['koor_y'])
  u = np.array(array_[i]['Vx'])
  v = np.array(array_[i]['Vy'])
  plt.quiver(x,y,u,v)
  plt.savefig(f'/content/images/images{i}.png')
```
